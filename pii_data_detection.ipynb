```python
from nltk.stem import PorterStemmer
from nltk.stem import RegexpStemmer
```

```python
ps = PorterStemmer()
rs = RegexpStemmer(r'ing|ed', min=3)
```

```python
def clean_doc(doc):
  # tokens = []
  tokens = list(map(ps.stem, [token for token in doc]))
  tokens = list(map(rs.stem, [token for token in tokens]))

  return tokens
```

```python
Xclean = [clean_doc(x) for x in tokens]  
```
