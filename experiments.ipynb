```python
def get_person_labels(chunked_sentences):
    for entity in chunked_sentences:
        if isinstance(entity, nltk.tree.Tree):
            if entity.label() == 'PERSON':
                print(' '.join([word for word, _ in entity.leaves()]))
```

```python
# ne_chunk(pos_tag(word_tokenize(text[0])))
tagged_words = pos_tag(tokens[0])
named_entities = ne_chunk(tagged_words)
get_person_labels(named_entities)
```
